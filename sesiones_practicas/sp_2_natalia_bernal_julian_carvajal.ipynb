### <span style="color:#2F749F;"><strong>Ejercicio 1: Cálculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528310` (debes reemplazarlo con los últimos 6 dígitos de tu documento de identificación).  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:  
   - Media teórica: `(base % 10) + 2`  
   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

image.png

```python
# Importamos librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Convertir cedula_1 a entero para usar como semilla
cedula_1 = int(cedula_1)
np.random.seed(cedula_1)

# Simular distribución normal
media_teorica = (cedula_1 % 10) + 2
desviacion_estandar = 2
n = 30

# Distribución
datos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=n)

# Reemplazar valores negativos por cero
datos[datos < 0] = 0

print(f"Ingresos generados (primeros 5): {datos[:5]}")
print(f"Media teórica usada: {media_teorica}")
```

Ahora que tenemos los datos de ingresos, vamos a calcular las medidas de tendencia central (media, mediana, moda) y a graficarlos como se solicita en el Ejercicio 1.

```python
# Cálculo de la Media
media = np.mean(datos)

# Cálculo de la Mediana
mediana = np.median(datos)

# Cálculo de la Moda
# Al ser una variable numérica el análisis de la mota no tiene sentido, sin embargo se deja por fines académicos
moda = stats.mode(datos)[0]

print(f"Media de los ingresos: {media:.2f}")
print(f"Mediana de los ingresos: {mediana:.2f}")
print(f"Moda de los ingresos: {moda}")
```

```python
# Gráfica de los resultados
plt.figure(figsize=(10, 6))
sns.histplot(datos, kde=True, color='skyblue', bins=10)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda}')
plt.title('Distribución de Ingresos y Medidas de Tendencia Central')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Ejercicio 1:

- **Media:** El valor de la media (línea roja) representa el promedio de los ingresos. Nos indica el valor central si todos los ingresos se distribuyeran equitativamente.

- **Mediana:** La mediana (línea verde) es el valor que divide la muestra en dos mitades iguales: el 50% de los ingresos son menores o iguales a este valor, y el 50% son mayores o iguales. Es menos sensible a valores atípicos que la media.

- **Moda:** La moda (línea morada) es el valor que aparece con mayor frecuencia en la muestra. En distribuciones continuas o con pocos datos, puede haber múltiples modas o no ser tan representativa.

Al observar el gráfico, la proximidad de la media, mediana y moda nos da una idea de la simetría de la distribución. Si están muy cerca, la distribución es probablemente simétrica. Las diferencias indican asimetría en los datos.

- **Desviación estándar:** Es la raíz cuadrada de la varianza e indica la dispersión promedio de los datos en las mismas unidades que la variable original. ### <span style="color:#2F749F;"><strong>Ejercicio 5: Análisis de correlación</strong></span>

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

image.png

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# Semilla usando los últimos 6 dígitos de la cédula
semilla = 419475
np.random.seed(semilla)

# Generar dos variables con distribución normal
ingresos = np.random.normal(loc=5000, scale=800, size=100)
gastos = ingresos * 0.7 + np.random.normal(loc=0, scale=300, size=100)

# Crear DataFrame
df = pd.DataFrame({
    "Ingresos": ingresos,
    "Gastos": gastos
})

# Calcular correlaciones
pearson = df["Ingresos"].corr(df["Gastos"], method="pearson")
spearman, _ = spearmanr(df["Ingresos"], df["Gastos"])

print("Correlación de Pearson:", pearson)
print("Correlación de Spearman:", spearman)

# Graficar relación
plt.figure(figsize=(8,5))
sns.scatterplot(x="Ingresos", y="Gastos", data=df)
sns.regplot(x="Ingresos", y="Gastos", data=df, scatter=False, color="red")

plt.title("Relación entre Ingresos y Gastos")
plt.xlabel("Ingresos")
plt.ylabel("Gastos")
plt.show()
```

### **Interpretación del Ejercicio 5**

Los coeficientes de correlación obtenidos indican que existe una relación positiva fuerte entre las variables Ingresos y Gastos.

El coeficiente de Pearson (0.8949) es cercano a 1, lo que significa que hay una fuerte relación lineal positiva. Esto indica que, en general, cuando los ingresos aumentan, los gastos también tienden a aumentar.

Por otro lado, el coeficiente de Spearman (0.8814) también muestra una relación positiva fuerte, lo que sugiere que el orden o tendencia entre las variables es consistente: valores más altos de ingresos suelen corresponder a valores más altos de gastos.

La gráfica de dispersión confirma este comportamiento, ya que los puntos siguen una tendencia ascendente y la línea de regresión muestra claramente una pendiente positiva. Esto refuerza la idea de que existe una asociación significativa entre ambas variables.

En conclusión, el análisis bivariado evidencia que los gastos tienden a aumentar a medida que aumentan los ingresos, lo cual es coherente con un comportamiento económico esperado.

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

image.png

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Semilla con los últimos 6 dígitos de la cédula
semilla = 419475
np.random.seed(semilla)

# Simular muestra de 60 personas
niveles = ["Básico", "Medio", "Superior"]
internet = ["Sí", "No"]

nivel_educativo = np.random.choice(niveles, 60)
acceso_internet = np.random.choice(internet, 60)

df = pd.DataFrame({
    "Nivel educativo": nivel_educativo,
    "Acceso a internet": acceso_internet
})

# Tabla de contingencia
tabla = pd.crosstab(df["Nivel educativo"], df["Acceso a internet"])
print("Tabla de contingencia:")
print(tabla)

# Prueba Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)

print("\nChi-cuadrado:", chi2)
print("p-valor:", p)
print("Grados de libertad:", dof)

# Mapa de calor
plt.figure(figsize=(6,4))
sns.heatmap(tabla, annot=True, cmap="Blues", fmt="d")

plt.title("Relación entre Nivel Educativo y Acceso a Internet")
plt.xlabel("Acceso a Internet")
plt.ylabel("Nivel Educativo")

plt.show()
```

### **Interpretación del Ejercicio 6**

A partir de la simulación de 60 observaciones se construyó una tabla de contingencia entre las variables Nivel educativo y Acceso a internet. Posteriormente se aplicó la prueba de Chi-cuadrado de independencia con el objetivo de evaluar si existe una asociación estadísticamente significativa entre ambas variables.

El resultado de la prueba arrojó un p-valor de 0.0782, el cual es mayor que el nivel de significancia común de 0.05. Por lo tanto, no se rechaza la hipótesis nula de independencia, lo que indica que no hay evidencia estadística suficiente para afirmar que exista una asociación significativa entre el nivel educativo y el acceso a internet en esta muestra simulada.

Sin embargo, el mapa de calor permite observar algunas diferencias en las frecuencias entre categorías, aunque estas variaciones no son lo suficientemente fuertes como para considerarse estadísticamente significativas según la prueba aplicada.

En conclusión, aunque visualmente se observan ciertas variaciones en la distribución de las categorías, el análisis estadístico indica que dichas diferencias podrían deberse al azar dentro de la muestra analizada.

```python
# Varianza explicada
varianza = pca.explained_variance_

# Proporción de varianza explicada
PVE = pca.explained_variance_ratio_

# Varianza acumulada
PVE_acum = np.cumsum(PVE)

componentes_labels = [f'PC{i+1}' for i in range(len(PVE))]

df_varianza = pd.DataFrame({
    "Componente": componentes_labels,
    "Autovalor": varianza,
    "PVE": PVE,
    "PVE acumulada": PVE_acum
})

df_varianza.round(3)
```
