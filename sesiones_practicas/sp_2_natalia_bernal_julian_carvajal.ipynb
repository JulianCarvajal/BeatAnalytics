### <span style="color:#2F749F;"><strong>Ejercicio 1: Cálculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528310` (debes reemplazarlo con los últimos 6 dígitos de tu documento de identificación).  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:  
   - Media teórica: `(base % 10) + 2`  
   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

```

```python
import numpy as np

# Convertir cedula_1 a entero para usar como semilla
np.random.seed(cedula_1)
```

A continuación, generamos los ingresos simulados con una distribución normal, utilizando la `seed_value` establecida y los parámetros definidos en el Ejercicio 1 (media teórica y desviación estándar). Los valores negativos se reemplazarán por cero.

```python
media_teorica = (int(cedula_1) % 10) + 2
desviacion_estandar = 2
n = 30

# Generar ingresos con distribución normal
datos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=n)

# Reemplazar valores negativos por cero
datos[datos < 0] = 0

print(f"Ingresos generados (primeros 5): {ingresos[:5]}")
print(f"Media teórica usada: {media_teorica}")
```

Ahora que tenemos los datos de ingresos, vamos a calcular las medidas de tendencia central (media, mediana, moda) y a graficarlos como se solicita en el Ejercicio 1.

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Cálculo de la Media
media = np.mean(datos)

# Cálculo de la Mediana
mediana = np.median(datos)

# Cálculo de la Moda
# stats.mode retorna la moda y su frecuencia. [0] obtiene la moda. [0] es para manejar el caso de múltiples modas.
moda = stats.mode(datos)[0]

print(f"Media de los ingresos: {media:.2f}")
print(f"Mediana de los ingresos: {mediana:.2f}")
print(f"Moda de los ingresos: {moda}")
```

```python
# Gráfica de los resultados
plt.figure(figsize=(10, 6))
sns.histplot(datos, kde=True, color='skyblue', bins=5)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda}')
plt.title('Distribución de Ingresos y Medidas de Tendencia Central')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Ejercicio 1:

- **Media:** El valor de la media (línea roja) representa el promedio de los ingresos. Nos indica el valor central si todos los ingresos se distribuyeran equitativamente.

- **Mediana:** La mediana (línea verde) es el valor que divide la muestra en dos mitades iguales: el 50% de los ingresos son menores o iguales a este valor, y el 50% son mayores o iguales. Es menos sensible a valores atípicos que la media.

- **Moda:** La moda (línea morada) es el valor que aparece con mayor frecuencia en la muestra. En distribuciones continuas o con pocos datos, puede haber múltiples modas o no ser tan representativa.

Al observar el gráfico, la proximidad de la media, mediana y moda nos da una idea de la simetría de la distribución. Si están muy cerca, la distribución es probablemente simétrica. Las diferencias indican asimetría en los datos.

### <span style="color:#2F749F;"><strong>Ejercicio 2: Cálculo de medidas de dispersión</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza qué tan dispersos están los datos respecto a su media. Calcula:
1. Rango
2. Varianza
3. Desviación estándar
4. Coeficiente de variación
5. Genera una gráfica con los resultados obtenidos (ver imagen de ejemplo)

```python
# Cálculo del Rango
rango = np.max(datos) - np.min(datos)

# Cálculo de la Varianza
varianza = np.var(datos)

# Cálculo de la Desviación estándar
desviacion_estandar_calculada = np.std(datos)

# Cálculo del Coeficiente de variación
# Se usa la desviación estándar calculada y la media obtenida en el Ejercicio 1
coeficiente_variacion = (desviacion_estandar_calculada / media) * 100

print(f"Rango de los ingresos: {rango:.2f}")
print(f"Varianza de los ingresos: {varianza:.2f}")
print(f"Desviación estándar de los ingresos: {desviacion_estandar_calculada:.2f}")
print(f"Coeficiente de variación de los ingresos: {coeficiente_variacion:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(datos, kde=True, color='purple', bins=5)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(media + desviacion_estandar_calculada, color='orange', linestyle='dashed', linewidth=2, label=f'Media + 1 Std Dev: {(media + desviacion_estandar_calculada):.2f}')
plt.axvline(media - desviacion_estandar_calculada, color='orange', linestyle='dashed', linewidth=2, label=f'Media - 1 Std Dev: {(media - desviacion_estandar_calculada):.2f}')
plt.title('Distribución de Ingresos y Medidas de Dispersión')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Ejercicio 2:

- **Rango:** El rango nos indica la amplitud total de los datos. En este caso, es la diferencia entre el ingreso más alto y el más bajo en la muestra.

- **Varianza:** La varianza mide la dispersión promedio de cada punto de dato respecto a la media. Un valor alto indica que los datos están más dispersos.

- **Desviación estándar:** Es la raíz cuadrada de la varianza e indica la dispersión promedio de los datos en las mismas unidades que la variable original. Ayuda a entender cuánto se alejan los valores individuales de la media. En la gráfica, las líneas naranjas muestran una desviación estándar por encima y por debajo de la media.

- **Coeficiente de variación:** Este coeficiente es una medida relativa de la dispersión, expresada como porcentaje. Es útil para comparar la variabilidad entre conjuntos de datos con diferentes medias. Un coeficiente más alto indica una mayor variabilidad relativa.

Al observar el gráfico, podemos visualizar cómo la mayoría de los datos se agrupan alrededor de la media dentro del rango de una desviación estándar. Si la distribución es aproximadamente normal, aproximadamente el 68% de los datos deberían caer dentro de este rango.

### <span style="color:#2F749F;"><strong>Ejercicio 4: Cálculo de medidas de percentiles</strong></span>

A partir de la muestra generada en el Ejercicio 1, calcula los percentiles de tus datos. Calcula:
1. Percentiles
2. Deciles
3. Cuartiles
4. Grafica los anteriores percentiles calculados, sobre la distribución de tus datos (ver la imagen de ejemplo)

image.png

```python
# Estilo visual
sns.set_theme(style="whitegrid")

# Cálculo de percentiles clave
percentiles = {
    'P10': {'valor': np.percentile(datos, 10), 'color': 'red', 'linestyle': '-'},
    'P25 (Q1)': {'valor': np.percentile(datos, 25), 'color': 'orange', 'linestyle': '--'},
    'P50 (Mediana)': {'valor': np.percentile(datos, 50), 'color': 'green', 'linestyle': '-.'},
    'P75 (Q3)': {'valor': np.percentile(datos, 75), 'color': 'blue', 'linestyle': ':'},
    'P90': {'valor': np.percentile(datos, 90), 'color': 'purple', 'linestyle': (0, (5, 2))}
}

# Cálculo de deciles (D1 a D9)
deciles = {f'D{i}': np.percentile(datos, i * 10) for i in range(1, 10)}

# Visualización
plt.figure(figsize=(14, 7))
sns.histplot(datos, kde=True, color='skyblue', bins=20, edgecolor="None", alpha=0.6)

# Líneas verticales de percentiles
for etiqueta, props in percentiles.items():
    plt.axvline(props['valor'], color=props['color'], linestyle=props['linestyle'], linewidth=2)
    plt.text(props['valor'], plt.ylim()[1]*0.92, etiqueta,
             rotation=90, color=props['color'], fontsize=10,
             verticalalignment='center', horizontalalignment='right',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Líneas verticales de deciles
for etiqueta, valor in deciles.items():
    plt.axvline(valor, color='gray', linestyle='--', alpha=0.3)
    plt.text(valor, plt.ylim()[1]*0.65, etiqueta,
             rotation=90, color='gray', fontsize=8,
             verticalalignment='center', horizontalalignment='right')

texto = (
    "Percentiles → dividen en 100 partes\n"
    "Deciles → dividen en 10 partes (D1 a D9)\n"
    "Cuartiles → Q1, Q2, Q3 (0.25, 0.50, 0.75)\n")
plt.text(0.98, 0.95, texto, transform=plt.gca().transAxes,
         fontsize=11, verticalalignment='top', horizontalalignment='right',
         bbox=dict(boxstyle="round,pad=0.5", facecolor="#f9f9f9", edgecolor="gray"))

plt.title("Distribución Normal con Percentiles y Deciles", fontsize=15, fontweight='bold')
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```
